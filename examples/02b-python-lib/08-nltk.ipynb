## NLTK токенайзер

```python
# import nltk
# nltk.download('punkt')
```

```python
# NLTK токенайзер
from nltk.tokenize import sent_tokenize as nltk_sentence_split
from nltk.tokenize import word_tokenize as nltk_tokenize_word

text_ = [ 
    nltk_tokenize_word(s,language='russian') # разбиваем предложения на слова
    for s in nltk_sentence_split(text,language='russian') # режем текст на отдельные предложения
]

display( len(text_) )
text_
```

```python
# собираем словарь
words = sorted( set.union( *[ set(s) for s in text_ ] ) )
display(len(words))
words
```

NLTK обеспечивает разделение на отдельные предложения   
и более качественную токенизацию чем простое деление по пробелам,

## Грамматический анализатор

```python
# !pip3 install svgling
```

```python
from nltk import CFG # Context-free grammar

# зададим грамматику в БНФ
grammar = CFG.fromstring("""
S -> OLT
OLT -> OBJ | OBJ UN OLT
OBJ ->  ADJ NP
NP -> 'сок' | 'морс' | 'компот'
ADJ -> 'апельсиновый' | 'вишнёвый' | 'абрикосовый' | 'томатный'
UN -> ',' |'и'
""")

# display( grammar ) 
display( grammar.start() ) # начальный нетерминал
display( grammar.productions()  ) # продукции (правила вывода)
```

```python
from nltk.tokenize import word_tokenize as nltk_tokenize_word
from nltk import ChartParser as Parser

# выпоняем грамматический разбор
sentence = 'вишнёвый сок, абрикосовый компот, апельсиновый морс и томатный сок'

tokens = nltk_tokenize_word(sentence)

for tree in Parser(grammar).parse(tokens):  
    # tree.pretty_print()
    display(tree)
```
